```python
train_dataset = datasets.MNIST(root = './', train = True, transform = transforms.ToTensor(), download = True)
test_dataset = datasets.MNIST(root = './', train = False, transform = transforms.ToTensor(), download = True)

train_loader = DataLoader(dataset = train_dataset, batch_size =32, shuffle = True)
test_loader = DataLoader(dataset = test_dataset, batch_size =32, shuffle = False)

#train_dataset.__dict__
train_loader.__dict__
```

```python
class model(nn.Module):
    def __init__(self):
        super(model, self).__init__()
        self.l1 = nn.Linear(784,520)
        self.l2 = nn.Linear(520,320)
        self.l3 = nn.Linear(320,240)
        self.l4 = nn.Linear(240,120)
        self.l5 = nn.Linear(120,10)
    
    def forward(self,x):
        x = x.view(-1,784)# n,1,28,28 -> n,784
        x = F.relu(self.l1(x))
        x = F.relu(self.l2(x))
        x = F.relu(self.l3(x))
        x = F.relu(self.l4(x))
        x = F.relu(self.l5(x))
        return x
    
```

```python
my_net = model()
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(my_net.parameters(),lr=0.01,momentum=0.5)
#def train(epoch):
#    my_net.train()
for batch_idx, (data,target) in enumerate(train_loader):
    data, target = Variable(data), Variable(target)
    optimizer.zero_grad()
    output = my_net(data)
    loss = criterion(output,target)
    if batch_idx%100 ==0:
        print(batch_idx, loss)
    loss.backward()
    optimizer.step()
    
```

```python
 %matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import datasets, transforms
from  torch.autograd import Variable
from torch.utils.data import Dataset, DataLoader

import numpy as np

import torchvision     # data set 불러오기위해
import torchvision.transforms as transforms
```

```python
class inception(nn.Module):
    def __init__(self, in_c):
        super(inception, self).__init__()
    
        self.branch1 = nn.Conv2d(in_c,24,1)
        
        self.branch2 = nn.Conv2d(in_c,24,1)
        
        self.branch3_1 = nn.Conv2d(in_c,16,1)
        self.branch3_2 = nn.Conv2d(16,24,5,1,2)
        
        self.branch4_1 = nn.Conv2d(in_c,16,1)
        self.branch4_2 = nn.Conv2d(16,24,3,1)
        self.branch4_3 = nn.Conv2d(24,24,3,1,2)
        
    def forward(self,x):
        branch1 = F.avg_pool2d(x,3,1,1)
        print("avg_pool: ",branch1.size())
        branch1 = self.branch1(branch1)
        print("branch1: ",branch1.size())
        
        branch2 = self.branch2(x)
        print("branch2: ",branch2.size())
        
        branch3 = self.branch3_1(x)
        branch3 = self.branch3_2(branch3)
        print("branch3: ",branch3.size())
        
        branch4 = self.branch4_1(x)
        branch4 = self.branch4_2(branch4)
        branch4 = self.branch4_3(branch4)
        print("branch4: ",branch4.size())
        
        outputs = [branch1,branch2,branch3,branch4]
        #print("outputs: ",output)
        
        return torch.cat(outputs,1)
        
```

```python
class net1(nn.Module):
    def __init__(self):
        super(net1,self).__init__()
        self.con1 = nn.Conv2d(1,10,5)
        self.con2 = nn.Conv2d(96,20,5)
        
        self.incept1 = inception(in_c = 10)
        self.incept2 = inception(in_c = 20)
        
        self.mp = nn.MaxPool2d(2)
        self.fc = nn.Linear(1536,10)
        
    def forward(self,x):
        
        x = F.relu(self.mp(self.con1(x)))
        print(x.size())
        x = self.incept1(x)
        print(x.size())
        x = F.relu(self.mp(self.con2(x)))
        print(x.size())
        x = self.incept2(x)
        print(x.size())
        x = x.view(x.size(0),-1)
        print(x.size())
        x = self.fc(x)
        print(x.size())
        return x
```

```python
my_module1 = net1()
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(my_module1.parameters(),lr=0.01,momentum=0.5)

#for epoch in range(1):
for i,(data, target) in enumerate(train_loader):
    data, target = Variable(data), Variable(target)
    optimizer.zero_grad()
    output = my_module1(data)
    loss = criterion(output,target)
    
    if i%100== 0:
        print(i, loss)
    
    loss.backward()
    optimizer.step()
```
