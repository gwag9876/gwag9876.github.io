```python
# autograd 와 backpropagation
# autograd 기능을 쓰기 위해서는 Variable을 사용해야한다

import torch
from torch.autograd import Variable
a = torch.rand(5)
a = Variable(a) # Variable은 data, grad, grad_fn 으로 이루어져 있다.
# data - tensor 형태의 data가 담김
# grad - data가 거쳐온 layer에 대한 미분값이 축적됨
# grad_fn - 미분 값을 계산한 함수에 대한 정보
```

```python
a = torch.ones(2,2)
print(a)
a = Variable(a, requires_grad = True)
print(a)
```

```python
print("----a.data----")
print(a.data)
print("----a.grad----")
print(a.grad)
print("----a.grad_fn----")
print(a.grad_fn)
```

```python
# torch.nn   vs  torch.nn.functional

# torch.nn.conv2d(input_channels, output_channels = filter의 갯수, kernel_size = filter 크기,
#                 stride= 1, peddding= 0, dilation= 1, groups= 1, bias= True = 필터 한장의 bias를 몇으로?)
# weight를 자동으로 선언된다. 따로 선언하지 않아도 된다.

# torch.nn.functional.conv2d(input,weight - 외부에서 만든 filter를 직접 넣어줘야 한다,
#                            bias= None, stride= 1, pedding= 0, dilation= 1, groups= 1)
#
import torch 
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Variable

input = torch.ones(1,1,3,3)
filter = torch.ones(1,1,3,3)

input = Variable(input, requires_grad = True)
print(input)
```

```python
filter = Variable(filter)
out = F.conv2d(input, filter)
print(out)
```

```python
out.backward()
print(out.grad_fn)
print(input.grad)
```

```python
import torch
import torch.nn as nn
from torch.autograd import Variable

input = torch.ones(1,1,3,3)
input = Variable(input, requires_grad = True)
print("------input------")
print(input)
func = nn.Conv2d(1,1,3) # input ch, output ch, size = filter 크기 3 by 3 -> 3으로 표기
print("\n","------func.weight = filter weight------")
print(func.weight)

out = func(input) # 자동으로 bias 값을 가지고 있기 때문에 w *1 의 합과 다르다.
print("\n","------out------")
print(out)
print("\n","------out.grad = back ward 전------")
print(input.grad)
out.backward()
print("\n","------out.grad = back ward 후------")
print(input.grad)
```

```python
class model(nn.Module):
    def __init__(self):
        super(model, self).__init__()
        self.net1 = nn.Conv2d(3,5,5)  # 내가 쓸 네트워크 종류
        self.net2 = nn.Conv2d(5,10,5)
        
    def forward(self,x):
        x = self.net1(x)  
        x = self.net2(x)
        return x
    
imgs = 0
for n,(img, labels) in enumerate(trainloader):
    print(n, img.shape, labels.shape)
    imgs = img
    break
    
mynet = model()
out = mynet(Variable(imgs))
print(out.shape)
```
