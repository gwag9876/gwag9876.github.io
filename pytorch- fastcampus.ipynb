```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.init as init
import torchvision.datasets as dset
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from torch.autograd import Variable
```

```python
mnist_train = dset.MNIST("./",train = True,transform = transforms.ToTensor(),download = True)
mnist_test = dset.MNIST("./",train = False,transform = transforms.ToTensor(),download = True)
```

```python
print(mnist_train.__getitem__(0)[1], mnist_train.__len__())
mnist_test.__getitem__(0)[1], mnist_test.__len__()
```

```python
train_loader = DataLoader(mnist_train,batch_size= 16,shuffle= True,num_workers=2,drop_last = True)
test_loader = DataLoader(mnist_test,batch_size= 16,shuffle= False,num_workers=2,drop_last = True)

train_loader.__dict__
```

```python
class Linear(nn.Module):
    def __init__(self):
        super(Linear,self).__init__()
        self.layer = nn.Sequential(
            nn.Linear(784,300),
            nn.ReLU(),
            nn.Linear(300,100),
            nn.ReLU(),
            nn.Linear(100,10),
            nn.ReLU())
        
    def forward(self,x):
        x = x.view(16,-1)
        x = self.layer(x)
        return x

        
```

```python
model = Linear().cuda()
```

```python
loss_f = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(),lr = 0.002)
```

```python
for i in range(10):
    for j, (image,label) in enumerate(train_loader):
        image, label = Variable(image).cuda(), Variable(label).cuda()
        
        optimizer.zero_grad()
        output = model(image)
        loss = loss_f(output,label)
        loss.backward()
        optimizer.step()
        
        if j %1000 ==0:
            print(loss)
```

```python
correct = 0
total = 0

for image,label in test_loader:
    image = Variable(image).cuda()
    label = Variable(label).cuda()
    
    output = model(image)
    _,predict = torch.max(output,1)
    
    total += label.size(0)
    correct += (predict == label).sum()
    
print("Accuracy : {}".format(100*correct/total))
```

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.init as init
import torchvision.datasets as dset
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from torch.autograd import Variable
```

```python
cifar_train = dset.CIFAR10("./",train = True,transform = transforms.ToTensor(),download = True)
cifar_test = dset.CIFAR10("./",train = False, transform = transforms.ToTensor(),download = True)
```

```python
print(cifar_train.__getitem__(0)[0].size(), cifar_train.__len__())
cifar_test.__getitem__(0)[0].size(), cifar_test.__len__()
```

```python
train_loader = DataLoader(cifar_train,256,shuffle = True, num_workers = 2,drop_last = True)
test_loader = DataLoader(cifar_test,256,shuffle= False, num_workers= 2,drop_last = True)
```

```python
class cnn1(nn.Module):
    def __init__(self):
        super(cnn1,self).__init__()
        self.layer = nn.Sequential(
            nn.Conv2d(3,16,3,padding=1), #flter size 3에 stride=1, padding=1 이면 사이즈 그대로
            nn.ReLU(),
            nn.Conv2d(16,32,3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2,2),
        
            nn.Conv2d(32,64,3,padding=1),
            nn.ReLU(),
            nn.Conv2d(64,128,3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2,2),
            
            nn.Conv2d(128,256,3,padding=1),
            nn.ReLU(),
            nn.Conv2d(256,256,3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2,2)
            )
        
        self.fc = nn.Sequential(
            nn.Linear(4*4*256,200),
            nn.ReLU(),
            nn.Linear(200,10)
            )
        
    def forward(self,x):
        x = self.layer(x)
        x = x.view(256,-1)
        x = self.fc(x)
        return x
```

```python
model3 = cnn1().cuda()
```

```python
loss_f = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model3.parameters(),lr =0.0002 )
```
